DiGraph for "Market States as Markov States

```python
import matplotlib.pyplot as plt
import networkx as nx
from IPython.display import Image

# Respective probabilities
nodes = ['Bull', 'Stagnant', 'Bear', 'Absorb_Bull', 'Absorb_Stagnant']
edges = [('Bull', 'Stagnant', {'probability': 0.3}),
         ('Bull', 'Bear', {'probability': 0.2}),
         ('Stagnant', 'Bear', {'probability': 0.4}),
         ('Bear', 'Absorb_Bull', {'probability': 0.1}),
         ('Bear', 'Absorb_Stagnant', {'probability': 0.1}),
         ('Stagnant', 'Absorb_Stagnant', {'probability': 0.2}),
         ('Stagnant', 'Absorb_Bull', {'probability': 0.1}),
         ('Bull', 'Absorb_Bull', {'probability': 0.1}),
         ('Bull', 'Absorb_Stagnant', {'probability': 0.1})]

# Creating a directed graph
G = nx.DiGraph()

# Adding nodes and edges to the graph
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# Extracting edge labels for probabilities
edge_labels = {(edge[0], edge[1]): f"Prob: {edge[2]['probability']}" for edge in edges}

# Plotting the graph
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=2000, font_size=10, font_weight='bold', edge_color='black', arrowsize=20)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='black', font_size=8)
ax = plt.gca()
for node in nodes:
    (x, y) = pos[node]
    ax.add_patch(plt.Circle((x, y), 0.15, color='white', ec='black', zorder=0))

plt.title('MARKET STATES AS MARKOV STATES')
plt.show()
plt.savefig('markov.png')

# Displaying the saved image
Image(filename='markov.png')
```
